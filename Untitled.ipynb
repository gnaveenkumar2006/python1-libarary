```python
import pandas as pd
df=pd.read_excel("aws.xlsx")

maxin=df["Quantity"].max()
minin=df["Quantity"].min()
df["normalization"]=(df["Quantity"] -minin) / (maxin - minin)
print(df[["Quantity","normalization"]])
```

```python
import pandas as pd
df=pd.read_excel("aws.xlsx")
mean=df["Quantity"].mean()
stand=df["Quantity"].std()
df["standardization"]=(df["Quantity"] - mean)/(stand)
print(df[["Quantity","standardization"]])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_excel("aws.xlsx")
df=pd.DataFrame(data)
plt.figure(figsize=(10,6))

sns.scatterplot(x="Product_Name",y="Total_Sales", color="orange",hue="Total_Sales",palette="cool", data=df)
plt.xlabel("Name")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1: Read your Excel file
df = pd.read_excel("aws.xlsx")

# Step 2: Filter rows that contain the word "South" (case-insensitive) in any column
south_rows = df.apply(lambda x: x.astype(str).str.contains('South', case=False, na=False)).any(axis=1)
south_data = df[south_rows]

# Step 3: Combine all text from these rows into one string
text = " ".join(south_data.astype(str).values.flatten())

# Step 4: Generate the WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Step 5: Display the WordCloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
